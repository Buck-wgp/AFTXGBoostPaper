```python
import sqlite3
from   matplotlib import pyplot as plt
import numpy as np
```

```python
def get_connection(table_name):
    conn = sqlite3.connect(table_name)
    cursor_obj = conn.cursor()
    return cursor_obj
```

```python
def get_trial_values(cursor_obj):
    trial_values = []
    cursor_obj.execute("SELECT * from trials")
    rows = cursor_obj.fetchall()
    n_trials = len(rows)
    for i in range(n_trials):
        trial_values.append(rows[i][3])
    trial_values = list(map(lambda x: np.nan if x==None else x,trial_values))
    return trial_values
```

```python
def get_param_values(cursor_obj):
    
    trial_values = []
    cursor_obj.execute("SELECT * from trial_params")
    rows = cursor_obj.fetchall()
    n_param   = len(rows)
    eta       = []
    max_depth = []
    
    for i in range(0,n_param,2):
        eta.append(rows[i][3])
        max_depth.append(rows[i+1][3])
        
    return eta,max_depth
```
